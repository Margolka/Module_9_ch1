```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
genres = pd.read_csv('tmdb_genres.csv')
genres['genres']=genres['genres'].fillna('unclassified')
genres.rename(columns={ genres.columns[0]: "genre_id" },inplace=True)
genres
```

```python
movies = pd.read_csv('tmdb_movies.csv')
movies.head()
```

```python
Q75=movies['vote_count'].quantile(0.75)
movies[movies['vote_count']>Q75].sort_values(by='vote_average',ascending=False).head(10)
```

2. Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 (włącznie) do 2016 roku (włącznie). Następnie na tej podstawie stwórz wykres, w którym średnie przychody są wykresem kolumnowym, a średni budżet wykresem liniowym na tych samych osiach. Sformatuj odpowiednio oś X oraz oś Y. Dodaj tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi. Przykład wykresu widoczny poniżej:

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies['release_date']=movies['release_date'].dt.year
mask = (movies['release_date'] >= 2010) & (movies['release_date'] <= 2016)
movies_by_date=movies.loc[mask].groupby('release_date')[['budget', 'revenue']].mean()
```

```python
movies_by_date=movies_by_date.reset_index()
def million(x, pos):
        return '{:2.1f}M'.format(x*1e-6)

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
formatter = plt.FuncFormatter(million)
axes.yaxis.set_major_formatter(formatter)

axes.plot(movies_by_date['release_date'],movies_by_date['budget'],label='budget',color = 'red')
axes.bar (movies_by_date['release_date'],movies_by_date['revenue'],label='revenue')
axes.legend(loc=(1.05,0.85))
axes.set_title('Średni przychód i budżet filmu w latach 2010-2016')
```
